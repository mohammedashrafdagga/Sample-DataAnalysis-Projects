```python
# importing libaray
import pandas as pd 
```

```python
# reading data
df = pd.read_csv('dateset/LondonHousingData.csv')

# looking a heading data
df.head()
```

```python
# look a count => non null cell for each columns 
df.count()
```

```python
# looking null value cell 
df.isnull().sum()
```

```python
# using Seaborn to show null value cell in columns
%matplotlib notebook 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.heatmap(df.isnull())
plt.show()
plt.subplots_adjust(bottom=0.25)
```

```python
# convert data type of columbn date to _datetime
# looking a type 
type(df['date'][0])
```

```python
# convert to pd.to_datetime
df['date'] = pd.to_datetime(df['date'])

df.dtypes
```

```python
df['date'][0]
```

```python
# add new columns name is Year contains only Year
df['year'] = df['date'].dt.year
# looking a new dataframe 
df.head()
```

```python
# add new columns month
df['month'] = df['date'].dt.month

df.head()
```

```python
# remove columns year and month 
df = df.drop(columns=['year','month'], axis=1)

# looking in df
df.head()
```

```python
# Show all records with no_of_crimes equal 0 and how many such records are there?
df[df['no_of_crimes'] == 0]
```

```python
"number of records no_of_crimes equal 0 is: {}".format(len(df[df['no_of_crimes'] == 0]))
```

```python
# Which Max and min average_price per year in england
print('max average per year is: {}'.format(df[df['area'] == 'england'].groupby(df['date'].dt.year)['average_price'].max()))
```

```python
print('min average per year is: {}'.format(df[df['area'] == 'england'].groupby(df['date'].dt.year)['average_price'].min()))
```

```python
# What the max and min no_of_cimes per area
print('max no_of_crimes per area is: {}'.format(df.groupby('area')['no_of_crimes'].max()))
```

```python
print('min no_of_crimes per area is: {}'.format(df.groupby('area')['no_of_crimes'].min()))
```

```python
# Show to total count of records of each area, where avrage is less than 100000
df[df['average_price'] < 100000]['area'].value_counts()
```
