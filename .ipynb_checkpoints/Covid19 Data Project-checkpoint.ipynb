```python
# import libaray
import pandas as pd
```

```python
df = pd.read_csv('dateset/covid_19_data.csv')

# looking a heading data
df.head()
```

```python
# null value
df.isnull().sum()
```

```python
# data shape
df.shape
```

```python
# coutn function return number non_null value cell for each columns
df.count()
```

```python
# cleaing columns name
# remove spaces and lowercase for all letter 
# strip remove space from word
# lower => conver letter word to lower case
df.columns = [x.strip().lower() for x in df.columns]
# print df.columns
df.columns
```

```python
# now return unique region columns value
df['region'].unique()
```

```python
# return unqiue state
df['state'].unique()
```

```python
# In region US
df[df['region'] == 'US'].head()
```

```python
# looking a shape in this df
df[df['region'] == 'US'].shape
```

```python
df[df['region'] == 'US']['state'].unique()
```

```python
# Specfic state from US 
df[(df['region'] == 'US') & (df['state'] == 'Florida')]
```

```python
# we create a list from first 10 Us State
df[df['region'] == 'US']['state'][:10]
```

```python
states = df[df['region'] == 'US']['state'][:10].tolist()

states
```

```python
# now plot sccater 
# improt matplot
%matplotlib notebook 
import matplotlib.pyplot as plt 
```

```python
# create confirmed list from first 
confirm = []
for stat in states:
    confirm.append(df[(df['region'] == 'US') & (df['state'] == stat)]['confirmed'].values[0])

confirm
```

```python
# now plloting
fig, ax = plt.subplots()


plt.xlabel("State")
plt.ylabel("Confirmed Cases")
plt.scatter(states,confirm)
plt.show()
```

```python
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
plt.subplots_adjust(bottom=0.35)  
```

```python
# for other region , for examplr Canda
df[df['region'] == 'UK']
```

```python
states = list(df[df['region'] == 'UK']['state'][1:].values)
confirm = []
for stat in states:
    confirm.append(df[(df['region'] == 'UK') & (df['state'] == stat)]['confirmed'].values[0])

confirm
```

```python
fig, ax = plt.subplots()


plt.xlabel("State in UK")
plt.ylabel("Confirmed Cases")
plt.scatter(states,confirm)
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
plt.subplots_adjust(bottom=0.35)  
plt.show()
```
